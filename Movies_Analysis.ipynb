## Limpeza dos Dados e Feature Engineering

```python
df = df.rename(columns={'title': 'Title'})
df = df.rename(columns={'overview': 'Synopsis'})
df = df.rename(columns={'popularity': 'Popularity'})
df = df.rename(columns={'vote_average': 'Rating'})
df = df.rename(columns={'vote_count': 'Vote_Count'})
df = df.rename(columns={'genre_ids': 'Genre_Ids'})
df = df.rename(columns={'release_date': 'Release_Date'})
```

```python
df.head()
```

```python
# Criar uma coluna com a quntidade de generos em cada linha
if df['Genre_Ids'].apply(type).iloc[0] == str:
    df['Genres_Count'] = df['Genre_Ids'].apply(lambda x: len(eval(x)))
elif df['Genre_Ids'].apply(type).iloc[0] == list:
    df['Genres_Count'] = df['Genre_Ids'].apply(len)
else:
    print("Os valores em 'Genre_Ids' estão em um formato inesperado.")
```

```python
df.head()
```

```python
# Filme com maior numero de generos.
maior_valor = df['Genres_Count'].max()
print("O maior valor na coluna 'Genres_Count' é:", maior_valor)
```

```python
df.info()
```

```python
df['Genre_Ids'] = df['Genre_Ids'].astype(str)
```

```python
df['Genre_Ids'] = df['Genre_Ids'].str.strip('[]')
```

```python
df.head()
```

```python
# Função para reduzir a lista de gêneros a três
def limitar_generos(Genre_Ids, limite=3):
    # Dividir a string por vírgulas para obter uma lista de gêneros
    lista_generos = Genre_Ids.split(', ')
    # Selecionar até 'limite' gêneros da lista
    generos_limitados = lista_generos[:limite]
    # Juntar os gêneros de volta em uma string com vírgulas
    return ', '.join(generos_limitados)

# Aplicar a função 'limitar_generos' à coluna 'Genre_Ids'
df['Genre_Ids'] = df['Genre_Ids'].apply(limitar_generos)

# Exibir as primeiras linhas do DataFrame para verificar o resultado
df.head()
```

```python
def separar_generos(Genre_Ids):
    # Dividir a string em uma lista de gêneros
    generos = Genre_Ids.split(', ')
    num_generos = len(generos)

    # Criar uma lista para armazenar os gêneros ou "ausente"
    resultado = []

    # Adicionar os três primeiros gêneros ou "ausente" se não houver três gêneros
    for i in range(3):
        if i < num_generos:
            resultado.append(generos[i])
        else:
            resultado.append('Ausente')

    return resultado

df[['Genre_1', 'Genre_2', 'Genre_3']] = df['Genre_Ids'].apply(separar_generos).apply(pd.Series)
```
