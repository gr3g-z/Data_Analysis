```python
#libs
import pandas as pd
import requests
```

```python
response = requests.get('https://api.themoviedb.org/3/movie/popular?api_key=fbc09e0a7e2ec80427ca1105d67751ae&language=en-US&page=1')
response
```

```python
response.json()['results'] #API sempre retorna JSON 
```

```python
df = pd.DataFrame(response.json()['results'])[['id','title','genre_ids','overview','release_date','popularity','vote_average','vote_count']]
df # usando o JSON como DataFrame
```

```python
# cada consulta retorna so uma pagina logo é feito um LOOP para pegar quantas paginas quiser! TOTAL:37432 paginas
for i in range(1,501):
    response = requests.get('https://api.themoviedb.org/3/movie/popular?api_key=fbc09e0a7e2ec80427ca1105d67751ae&language=en-US&page={}'.format(i))
    data = pd.DataFrame(response.json()['results'])[['id','title','genre_ids','overview','release_date','popularity','vote_average','vote_count']]
    df = df.append(data, ignore_index = True)
```
